```python
import pandas as pd
import matplotlib.pyplot as plt 

lokali = pd.read_csv('csv_datoteke/lokali.csv',delimiter = ";")
ocene = pd.read_csv('csv_datoteke/ocene.csv',delimiter = ";")
jedi= pd.read_csv('csv_datoteke/jedi.csv',delimiter = ";")
posebnosti = pd.read_csv('csv_datoteke/posebnosti.csv',delimiter = ";")
```

```python
lokali_in_ocene = pd.merge(lokali, ocene, on= 'id')
lokali_in_ocene
```

```python
lokali_in_ocene.groupby('ocena').size()
```

```python
lokali_in_ocene.groupby('ocena').size().plot.bar()
```

```python
lokali_in_ocene.groupby('mesto')['ocena'].mean().sort_values(ascending=False)
```

```python
lokali_in_ocene.groupby('mesto')['doplacilo'].mean().sort_values(ascending=False)
```

```python
lokali_in_ocene.groupby('mesto')['doplacilo'].mean().plot(kind="bar")
```

```python
lokali_in_ocene.groupby('doplacilo')['ocena'].mean().plot()
```

```python
jedi.groupby('vrsta').size()
```

```python
posebnosti.drop(['id_lokala'], axis=1).apply(pd.Series.value_counts).iloc[0].plot.bar()
```

```python
posebnosti_num = posebnosti.replace(['Da','Ne'],[1,0])
posebnosti_num['vsota'] = posebnosti_num.drop(['id_lokala'], axis=1).sum(axis=1)

lokali_in_posebnosti = pd.merge(lokali, posebnosti_num, left_on= 'id', right_on='id_lokala')
lokali_in_posebnosti
```

```python
lokali_in_posebnosti.groupby('vsota')['doplacilo'].mean().plot(kind="bar")
plt.xlabel("Število pozitivnih lastnosti")
plt.show()
```

```python
meniji = jedi.groupby('id_lokala').count().reset_index().drop('vrsta', axis=1)
meniji['stevilo_menijev'] = meniji['jed']
meniji.drop('jed', axis=1)
lokali_in_meniji = pd.merge(lokali, meniji, left_on= 'id', right_on='id_lokala')
lokali_in_meniji
```
